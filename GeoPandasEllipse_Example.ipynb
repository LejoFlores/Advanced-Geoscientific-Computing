```python
import geopandas as gpd
from shapely.geometry import Point
from shapely.affinity import scale, rotate
```

```python
# 1. Define center
center_point = Point(565300, 4845700) # Bogus Basin Snotel site coords in UTM

# 2. Create a circle (using a large buffer to define the major axis length)
# Example: 1000m radius
circle = center_point.buffer(1000)

# 3. Scale it to an ellipse (x-axis (width) = 1000m, y-axis (height) = 500m)
# Scale factor: (scale_x, scale_y)
ellipse = scale(circle, 1, 0.5)

# 4. Rotate the ellipse (e.g., 45 degrees clockwise)
# Note: Positive degrees = counter-clockwise
oriented_ellipse = rotate(ellipse, -45)

# Convert to GeoSeries
gs = gpd.GeoSeries([oriented_ellipse])
```

```python
gs.plot(figsize=(10,10))
```
